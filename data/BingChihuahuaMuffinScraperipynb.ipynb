```python
from bing_image_downloader import downloader
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
dir = '/content/drive/MyDrive/image dataset'
```

```python
# Set the desired total number of images
total_images = 10000
# Set the limit per query
query_limit = 150  # Adjust as needed

# Set the limit per query
query_limit = 75

# Define the search query
search_query = 'blueberry muffin'

# Initialize variables
downloaded_images = 0

while downloaded_images < total_images:
    # Download images for the current query
    downloader.download(search_query, limit=query_limit, output_dir=dir, adult_filter_off=False)

    # Update variables
    downloaded_images += query_limit

print(f"Downloaded {total_images} images for '{search_query}' in total.")
```
